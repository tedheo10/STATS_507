```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures, SplineTransformer, StandardScaler, LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import KFold
from sklearn.neighbors import KNeighborsRegressor
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error


boston = pd.read_csv('boston.csv')
boston.dropna()
X = boston[['indus']].values
y = boston['nox'].values

pipeline_spline = Pipeline([
    ('spline_features', SplineTransformer(n_knots=4, degree=3, knots=np.array([0, 12, 17, 30]).reshape(-1,1))),
    ('scaler', StandardScaler()),
    ('regressor', LinearRegression())
])
pipeline_spline.fit(X, y)

pipeline_poly = Pipeline([
    ('polynomial_features', PolynomialFeatures(degree=12)),
    ('scaler', StandardScaler()),
    ('regressor', LinearRegression()) 
])
pipeline_poly.fit(X, y)

knn = KNeighborsRegressor(n_neighbors=13)
knn.fit(X,y)

kf = KFold(n_splits=4, shuffle=True, random_state=42)

models = [('Spline', pipeline_spline), ('Polynomial', pipeline_poly), ('KNN', knn)]
mse_store = {}

for name, model in models:
    mse_cv = []
    for train, test in kf.split(X):
        X_train, X_test = X[train], X[test]
        y_train, y_test = y[train], y[test]
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        mse_cv.append(mean_squared_error(y_test, y_pred))
    mse_store[name] = np.mean(mse_cv)

print("[MSE Report]")
for name, mse in mse_store.items():
    print(f'{name} : {mse:.5f}')

print("")
print(f'[The lowest MSE] : {min(mse_store, key=mse_store.get)}')
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 6))
axs[0].scatter(X, pipeline_spline.predict(X), color='red', alpha=0.1)
axs[0].set_xlabel('INDUS')
axs[0].set_ylabel('NOX')
axs[0].set_title('Spline')
axs[1].scatter(X, pipeline_poly.predict(X), color='blue', alpha=0.1)
axs[1].set_xlabel('INDUS')
axs[1].set_ylabel('NOX')
axs[1].set_title('Polynomial')
axs[2].scatter(X, knn.predict(X), color='green', alpha=0.1)
axs[2].set_xlabel('INDUS')
axs[2].set_ylabel('NOX')
axs[2].set_title('KNN')
plt.tight_layout()
plt.show()
```

**you can Even though the spline model has the highest error, its pattern is the easiest to explain to the public among all models. While statistical accuracy is an important factor for model selection, sometimes conveying the meaning to the audience is more crucial if the errors do not significantly impact predictions.**

```python
from pygam import LinearGAM, LogisticGAM, s, f
from sklearn.model_selection import GridSearchCV

X = boston[['dis', 'indus', 'rad']].values
y = boston['nox'].values

lam = [0.5, 1, 5, 10]
gam = LinearGAM(s(0) + s(1) + s(2))
gam.gridsearch(X, y, lam=lam)

print(f"The best lambda for 'dis' = {gam.lam[0][0]}, The best lambda for 'indus' = {gam.lam[1][0]}, The best lambda for 'rad' = {gam.lam[2][0]}")
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 6))
titles = ['DIS', 'INDUS', 'RAD']

for i, ax in enumerate(axs):
    XX = gam.generate_X_grid(term=i)
    pdep, confi = gam.partial_dependence(term=i, X=XX, width=.95)
    ax.plot(XX[:, i], pdep)
    ax.plot(XX[:, i], confi, c='r', ls='--')
    ax.set_title(f'Partial Dependence of {titles[i]}')
    ax.set_xlabel(titles[i])
    ax.set_ylabel('NOX')
    
plt.tight_layout()
plt.show()
```

**The estimated partial dependence of 'rad' does not fully represent the true relationship between 'rad' and 'nox'. The plot shows a wavy pattern, suggesting a high-degree polynomial relationship, which makes explaining the causal relationship between them difficult. Additionally, the confidence bands widen in certain areas, indicating high uncertainty, and suggesting possible correlations with other factors that influence 'nox'.**
